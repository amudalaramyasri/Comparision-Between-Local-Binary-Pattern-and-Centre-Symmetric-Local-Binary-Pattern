```python
import cv2
import numpy as np
import time

def binary_to_decimal(g):
    total_sum = 0
    k = 0
    sz = g.shape

    elements = []

    for j in range(sz[1] - 1):
        i = 0
        elements.append(g[i, j])
        k += 1

    for i in range(sz[0] - 1):
        j = sz[1] - 1
        elements.append(g[i, j])
        k += 1

    for j in range(sz[1] - 1, 0, -1):
        i = sz[0] - 1
        elements.append(g[i, j])
        k += 1

    for i in range(sz[0] - 1, 1, -1):
        j = 0
        elements.append(g[i, j])
        k += 1

    szele = len(elements)
    reversed_elements = elements[::-1]

    for j in range(szele):
        total_sum += reversed_elements[j] * (2 ** (szele - j - 1))

    return total_sum

def myfcn_LBP(g):
    sz = g.shape
    rows = sz[0]
    columns = sz[1]
    center = rows // 2

    result = np.zeros_like(g)

    for i in range(rows):
        for j in range(columns):
            if g[center, center] <= g[i, j]:
                result[i, j] = 1

    return result

start_time = time.time()
# Load an example image (replace with your own image)
originalImage = cv2.imread('lighthouse_512.BMP', cv2.IMREAD_GRAYSCALE)
newSize = (511, 511)
originalImage = cv2.resize(originalImage, newSize)

# Specify the amount of padding on all sides
topPad = 1
bottomPad = 1
leftPad = 1
rightPad = 1

# Define the value for padding (e.g., 0 for black, 255 for white)
paddingValue = 0

# Use the cv2.copyMakeBorder function to add padding
paddedImage = cv2.copyMakeBorder(originalImage, topPad, bottomPad, leftPad, rightPad, cv2.BORDER_CONSTANT, value=paddingValue)

sz = paddedImage.shape
lbpImage = np.zeros(originalImage.shape, dtype=np.uint8)

# Initialize a cell array to store the 3x3 windows
# Loop through the image and extract 3x3 windows
for i in range(1, sz[0] - 1):
    for j in range(1, sz[1] - 1):
        # Extract the 3*3 window centered at (i, j)
        window = paddedImage[i-1:i+2, j-1:j+2]

        ImgLBP = myfcn_LBP(window)  # calculates LBP binary code
        value = binary_to_decimal(ImgLBP)  # converts binary to decimal number
        lbpImage[i-1, j-1] = value

        
cv2.imwrite("lbp_image.png", lbpImage)
cv2.imshow("LBP Image", lbpImage)
cv2.waitKey(0)
cv2.destroyAllWindows()

end_time = time.time()
 
   
    
time_taken_lbp = end_time - start_time

print(time_taken_lbp)
```

```python
import cv2
import numpy as np

def myfcn_CSLBP(g):
    result = np.zeros(4, dtype=int)
    
    if g[1, 2] > g[1, 0]:
        result[0] = 1
    if g[0, 2] > g[2, 0]:
        result[1] = 1
    if g[0, 1] > g[2, 1]:
        result[2] = 1
    if g[0, 0] > g[2, 2]:
        result[3] = 1
    
    h = 0
    szele = result.shape[0]
    for j in range(szele):
        h += result[j] * 2**(szele - j - 1)
    
    return h

start_time = time.time()
# Load an example image (replace with your own image)
originalImage = cv2.imread('lighthouse_512.BMP', cv2.IMREAD_GRAYSCALE)
newSize = (510, 510)
originalImage = cv2.resize(originalImage, newSize)

# Specify the amount of padding on all sides
topPad = 1
bottomPad = 1
leftPad = 1
rightPad = 1

# Define the value for padding (e.g., 0 for black, 255 for white)
paddingValue = 0

# Use the cv2.copyMakeBorder function to add padding
paddedImage = cv2.copyMakeBorder(originalImage, topPad, bottomPad, leftPad, rightPad, cv2.BORDER_CONSTANT, value=paddingValue)

sz = paddedImage.shape

CSLbpImage = np.zeros(originalImage.shape, dtype=np.uint8)

# Initialize a cell array to store the 3x3 windows
# Loop through the image and extract 3x3 windows
for i in range(1, sz[0] - 1):
    for j in range(1, sz[1] - 1):
        # Extract the 3x3 window centered at (i, j)
        window = paddedImage[i-1:i+2, j-1:j+2]
        # Store the window in the cell array
        sum = myfcn_CSLBP(window)  # calculates CSLBP value for each 3x3 window
        CSLbpImage[i-1, j-1] = sum*255/15

cv2.imshow("CSLBP Image", CSLbpImage)
cv2.waitKey(0)
cv2.destroyAllWindows()

end_time = time.time()
 
   
    
time_taken_cslbp = end_time - start_time

print(time_taken_cslbp)
```
